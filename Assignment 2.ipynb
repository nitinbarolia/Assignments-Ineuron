**Ques 1- Create the below pattern using nested for loop in Python.**

```
n=int(input("Number of max star for pattern:"))

for i in range(0,n):
  for j in range(0,i+1):
      print('*',end='')
  print('\r') 
for i in range(n,0,-1):
  for j in range(0,i+1):
    print('*',end='')
  print('\r')
  
    

```
